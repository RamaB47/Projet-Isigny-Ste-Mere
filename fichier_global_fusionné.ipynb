```python
date_mois = {
    "Janvier": "01",
    "Février": "02",
    "Mars": "03",
    "Avril":"04",
    "Mai": "05",
    "Juin": "06",
    "Juillet": "07",
    "Août": "08",
    "Septembre": "09",
    "Octobre": "10",
    "Novembre": "11",
    "Décembre": "12"
}
```

```python
print(df_rename['Date'].unique())
```

```python
# 1. On sépare le mois et l'année en deux colonnes temporaires
df_rename['Date'] = df_rename['Date'].str.strip()
```

```python
df_split = df_rename['Date'].str.split(r'\s+', n=1, expand=True)
```

```python
df_rename['Nom_Mois'] = df_split[0]
df_rename['Annee'] = df_split[1]
```

```python
print(df_rename)
```

```python
print(df_rename.dtypes)
```

```python
ligne_drop = [0,1,2,3]
```

```python
df_rename.drop(axis=0, index = ligne_drop, inplace=True)
```

```python
print(df_rename)
```

```python
Month = {"Janvier": "01",
      "Février": "02",
      "Mars": "03",
      "Avril":"04",
      "Mai": "05",
      "Juin": "06",
      "Juillet": "07",
      "Août": "08",
      "Septembre": "09",
      "Octobre": "10",
      "Novembre": "11",
      "Décembre": "12"}
```

```python
df_rename["MM"] = df_rename['Nom_Mois'].map(Month)
```

```python
print(df_rename)
```

```python
df_rename["AAAAMM"]= df_rename["Annee"] + df_rename["MM"]
```

```python
print(df_rename)
```

```python
Col1 = ["Date","Nom_Mois", "Annee", "MM"]
```

```python
df_rename.drop(Col1, axis= 1, inplace=True)
```

```python
print(df_rename)
```

```python
from datetime import datetime
```

```python
df_rename['AAAAMM'] = pd.to_datetime(df_rename['AAAAMM'], format='%Y%m')
```

```python
print(df_rename)
```

```python
# On retransforme la vraie date en texte "YYYYMM"
df_rename['AAAAMM'] = df_rename['AAAAMM'].dt.strftime('%Y%m')
```

```python
print(df_rename)
```

```python
df_rename.to_csv("info_lait_final.csv")
```
