```python
import pandas as pd
```

```python

# 1. On charge les deux fichiers séparément (en spécifiant le bon séparateur, souvent ';' pour Météo France)
df1 = pd.read_csv('Meteo_50.csv', sep=';')
df2 = pd.read_csv('Meteo50-2.csv', sep=';')
```

```python
print(df1.columns.tolist())
```

```python
import pandas as pd


# 2. On les colle l'un sous l'autre (l'en-tête du 2e est géré automatiquement, pas de doublon !)
df_total = pd.concat([df1, df2], ignore_index=True)

# 3. On nettoie les espaces cachés dans les noms de colonnes
df_total.columns = df_total.columns.str.strip()

# 4. On applique votre filtre pour ne garder que vos 9 colonnes
colonnes_a_garder = ["NUM_POSTE", "NOM_USUEL", "RR", "AAAAMM", "NBRR", "TX","TN", "TM"]
df_final = df_total[colonnes_a_garder]

# On regarde le résultat
df_final.head()
```
